```python
def gcd_euclidean(a: int, b: int) -> int:
    """
    Знаходить найбільший спільний дільник двох чисел за допомогою алгоритму Евкліда.
    """
    if a <= 0 or b <= 0:
        raise ValueError("Числа повинні бути додатними.")
    print(f"Знаходимо НСД для чисел {a} і {b}")
    while b != 0:
        a, b = b, a % b
        print(f"Поточний стан: a = {a}, b = {b}")
    return a

# Приклади:
print("Приклад 1:")
print(f"Найбільший спільний дільник 16 і 36 – це {gcd_euclidean(16, 36)}")

print("\nПриклад 2:")
print(f"Найбільший спільний дільник 12 і 54 – це {gcd_euclidean(12, 54)}")
```
